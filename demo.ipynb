```python
# import the library
from surf import *
```

```python
# make font big
plt.rcParams.update({'font.size': 16})
```

```python
# visualize the surface's area element, gauss curvature, and mean curvature
def visualize(surf_fn, title=None, **kw):
  fig = plt.figure(figsize=(16,4))
  surface = Surface(surf_fn)
  surface.plot(show='area_element', ax=plt.subplot(131, projection='3d'), **kw)
  surface.plot(show='gauss_curvature', ax=plt.subplot(132, projection='3d'), **kw)
  surface.plot(show='mean_curvature', ax=plt.subplot(133, projection='3d'), **kw)
  plt.tight_layout()
```

## Plane

```python
def plane(x):
  u = x[0]
  v = x[1]
  return np.array([ u, v, np.zeros_like(u) ])

visualize(plane, over=[[-2,2],[-2,2]])
```

## Cylinder

```python
def cylinder(x):
  u = x[0]
  v = x[1]
  return np.array([ np.cos(u), np.sin(u), v ])

visualize(cylinder, over=[[0.001,2*np.pi],[-2,2]])
```

## Sphere, spherical coordinates

```python
def standard_sphere(x):
  u = x[0]
  v = x[1]
  return np.array([ 
    np.sin(u)*np.cos(v),
    np.sin(u)*np.sin(v),
    np.cos(u)
  ])

visualize(standard_sphere, over=[[0.001,np.pi],[0.001,2*np.pi]])
```

## Sphere, stereographic projection

```python
def stereo_sphere(x):
  u = x[0]
  v = x[1]
  return np.array([ 2*u, 2*v, u**2 + v**2 - 1 ]) / (1 + u**2 + v**2)

visualize(stereo_sphere, over=[[-2,2],[-2,2]])
```

## Catenoid

```python
def catenoid(x):
  U, V = x
  X = np.cosh(U)*np.cos(V)
  Y = np.cosh(U)*np.sin(V)
  return np.array([X,Y,U])

visualize(catenoid, over=[[-2,2],[0,2*np.pi]])
```

## Helicoid

```python
def helicoid(x):
  U, V = x
  X = U*np.cos(V)
  Y = U*np.sin(V)
  return np.array([X,Y,V])

visualize(helicoid, over=[[-2,2],[0,2*np.pi]])
```

## Möbius Strip (note reversed mean curvature)

```python
def mobius_strip(x):
  U, V = x
  X = np.cos(V) + U * np.sin(V/2) * np.cos(V)
  Y = np.sin(V) + U * np.sin(V/2) * np.sin(V)
  Z = U * np.cos(V/2)
  return np.array([X,Y,Z])

visualize(mobius_strip, over=[[-2,2],[0,2*np.pi]])
```
