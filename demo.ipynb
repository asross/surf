```python
plt.rcParams.update({'font.size': 16})
```

```python
def plot_grid(surface, **kw):
  surface.plot(show='area', ax=plt.subplot(131, projection='3d'), **kw)
  surface.plot(show='gauss_curvature', ax=plt.subplot(132, projection='3d'), **kw)
  surface.plot(show='mean_curvature', ax=plt.subplot(133, projection='3d'), **kw)
  plt.tight_layout()
```

```python
def stereo_projection(x):
  u = x[0]
  v = x[1]
  return np.array([ 2*u, 2*v, u**2 + v**2 - 1 ]) / (1 + u**2 + v**2)

sphere = Surface(stereo_projection)
fig = plt.figure(figsize=(16,4))
fig.suptitle('Sphere, stereographic projection', va='bottom', fontsize=20)
plot_grid(sphere, over=[[-2,2],[-2,2],50])
```

```python
def catenoid_(x):
  U, V = x
  X = np.cosh(U)*np.cos(V)
  Y = np.cosh(U)*np.sin(V)
  return np.array([X,Y,U])

catenoid = Surface(catenoid_)

fig = plt.figure(figsize=(16,4))
fig.suptitle('Catenoid', va='bottom', fontsize=20)
plot_grid(catenoid, over=[[-2,2],[0,2*np.pi]])
```

```python
def conformal1(x):
  U, V = x
  X = U
  Y = V
  Z = 2*U*V
  return np.array([X,Y,Z])
```

```python
def conformal2(x):
  U, V = x
  X = U
  Y = V
  Z = X**2 - Y**2
  return np.array([X,Y,Z])
```

```python
fig = plt.figure(figsize=(16,4))
fig.suptitle('$f(u,v) = (u, v, 2uv)$', va='bottom', fontsize=20)
plot_grid(Surface(conformal1), over=[[-3,3],[-3,3]])

fig = plt.figure(figsize=(16,4))
fig.suptitle('$f(u,v) = (u, v, u^2-v^2)$', va='bottom', fontsize=20)
plot_grid(Surface(conformal2), over=[[-3,3],[-3,3]])
```
